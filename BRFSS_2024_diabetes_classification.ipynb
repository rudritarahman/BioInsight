## Purpose

The purpose of this code notebook is to preprocess **BRFSS survey data 2024** into a useable format for prediction diabetes into **3 classes**:
1. No Diabetes
2. Pre-Diabetes
3. Diabetes (Type 2 only).

The dataset originally has **457,670 survey** responses and **301 features (columns)**, but based on diabetes disease research regarding factors influencing Type 2 diabetes disease, only selected features are included in this analysis.


**Link to Dataset:** [Diabetes Health Indicators Dataset](https://www.kaggle.com/datasets/rudritarahman/cdc-brfss-survey-data-2024)

## 3. Create Diabetes Labels

```python
target_df = main_df.copy()
```

```python
results = {
    "Condition": [
        "Yes and Type 1",
        "Yes and Type 2"
    ],
    "Count": [
        ((target_df["Diabetes_Status"] == 1) & (target_df["Diabetes_Type"] == 1)).sum(),
        ((target_df["Diabetes_Status"] == 1) & (target_df["Diabetes_Type"] == 2)).sum()
    ],
}

# Summary table
summary = pd.DataFrame(results).set_index("Condition")

# Percentage
total = len(target_df)
summary["Percent"] = round(summary["Count"] / total * 100, 2)

summary
```

```python
def assign_target(row):
    """
    Assign target labels for diabetes modeling:
    0 → No diabetes
    1 → Pre-diabetes
    2 → Type 2 diabetes
    "REMOVE" → Exclude from modeling (e.g., Type 1 or ambiguous)
    """

    # 1. Missing or refused Diabetes_Status → remove
    if row["Diabetes_Status"] in [7, 9] or pd.isna(row["Diabetes_Status"]):
        return np.nan

    # 2. Pregnant diabetes → classify as No diabetes
    if row["Diabetes_Status"] == 2 or row.get("Prediabetes_Status") == 2:
        return "No Diabetes"

    # 3. No diabetes
    if row["Diabetes_Status"] == 3:
        return "No Diabetes"

    # 4. Pre-diabetes
    if row["Diabetes_Status"] == 4 or row.get("Prediabetes_Status") == 4:
        return "Prediabetes"

    # 5. Diabetes_Status = 1 → diabetic
    if row["Diabetes_Status"] == 1:
        # Missing or refused Diabetes_Type → ambiguous → exclude
        if row["Diabetes_Type"] in [7, 9] or pd.isna(row["Diabetes_Type"]):
            return np.nan

        # Type 1 → remove from Type 2 modeling
        if row["Diabetes_Type"] == 1:
            return "REMOVE"

        # Type 2 → final class
        if row["Diabetes_Type"] == 2:
            return "Type 2 Diabetes"

    # Catch-all → exclude any other ambiguous cases
    return np.nan
```

```python
# "Target" class add
target_df["Target"] = target_df.apply(assign_target, axis=1)
```

```python
target_df["Target"].value_counts(normalize=True)
```

**Implications**

* A naïve classifier predicting only No Diabetes would get **>94%** accuracy, but it would completely fail to detect pre-diabetes or Type 2 diabetes.
* Both minority classes (1 and 2) are almost equal in proportion (**~2.8%**), but very small compared to the majority class.

```python
sns.set_style('whitegrid')
sns.set_palette('pastel')

def detect_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = series[(series < lower) | (series > upper)]
    return outliers, lower, upper

for col in numeric_features:
    data = outlier_df[col].dropna()

    # Calculate statistics
    Q1 = round(data.quantile(0.25), 4)
    Q3 = round(data.quantile(0.75), 4)
    IQR = round(Q3 - Q1, 4)
    mean_val = round(data.mean(), 4)
    median_val = round(data.median(), 4)
    min_val = round(data.min(), 4)
    max_val = round(data.max(), 4)

    # Detect outliers
    outliers, lower_bound, upper_bound = detect_outliers_iqr(data)

    print(f"\nStatistics for {col}:")
    print(f"  Min: {min_val}")
    print(f"  Q1: {Q1}")
    print(f"  Median: {median_val}")
    print(f"  Mean: {mean_val}")
    print(f"  Q3: {Q3}")
    print(f"  Max: {max_val}")
    print(f"  IQR: {IQR}")
    print(f"  Outliers: {len(outliers)} ({len(outliers)/len(data):.2%} of data)")

    # Plotting
    fig, axes = plt.subplots(1, 2, figsize=(14,5))

    # Boxplot with outliers highlighted
    sns.boxplot(x=data, ax=axes[0],
                boxprops=dict(facecolor='lightblue'),
                flierprops=dict(marker='o', color='red', markersize=5))
    axes[0].axvline(mean_val, color='green', linestyle='--', label=f'Mean: {mean_val:.2f}')
    axes[0].set_title(f'Boxplot for {col}', fontsize=14, weight='bold')
    axes[0].set_xlabel(col, fontsize=12)
    axes[0].legend()

    # Histogram with KDE
    sns.histplot(data, kde=True, bins=30, ax=axes[1], color='skyblue')
    axes[1].axvline(mean_val, color='green', linestyle='--', label=f'Mean: {mean_val:.2f}')
    axes[1].axvline(median_val, color='orange', linestyle='--', label=f'Median: {median_val:.2f}')
    axes[1].axvline(lower_bound, color='red', linestyle=':', label=f'Lower Bound: {lower_bound:.2f}')
    axes[1].axvline(upper_bound, color='red', linestyle=':', label=f'Upper Bound: {upper_bound:.2f}')
    axes[1].set_title(f'Histogram for {col}', fontsize=14, weight='bold')
    axes[1].set_xlabel(col, fontsize=12)
    axes[1].set_ylabel('Frequency', fontsize=12)
    axes[1].legend()

    plt.tight_layout()
    plt.show()
```

### Cramér's V (Nominal Features)

```python
# Cramér's V Function
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix, correction=False)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - (k-1)*(r-1)/(n-1))
    rcorr = r - (r-1)**2/(n-1)
    kcorr = k - (k-1)**2/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))
```

```python
def cramers_v_matrix(df, features):
    """Compute Cramér’s V matrix for all nominal features."""
    n = len(features)
    matrix = np.zeros((n, n))

    for i, col1 in enumerate(features):
        for j, col2 in enumerate(features):
            matrix[i, j] = cramers_v(df[col1], df[col2])

    return pd.DataFrame(matrix, index=features, columns=features)
```

## 8. Missing Value Check

```python
def missing_report(
    df,
    cmap="Reds",
    sort_by=None,
    ascending=False
):
    """
    Generate a styled missing-value report for any dataframe.

    Parameters:
    -----------
    df : pandas.DataFrame
        The dataframe to analyze.
    cmap : str
        Color map for background gradient (default: "Reds").
    sort_by : str
        Column to sort by ("Missing Count" or "Missing Percentage"), default: None.
    ascending : bool
        Sort order (default: False for descending).

    Returns:
    --------
    pandas.io.formats.style.Styler
        A styled report showing missing counts and percentages.
    """

    # No. of total rows
    total_rows = len(df)

    # Compute count and percentage
    missing_count = df.isna().sum()
    missing_percentage = ((missing_count / total_rows) * 100).round(2)

    # Build summary table
    missing_summary = pd.DataFrame({
        "Missing Count": missing_count,
        "Missing Percentage": missing_percentage
    })

    # Sorting if requested
    if sort_by in ["Missing Count", "Missing Percentage"]:
        missing_summary = missing_summary.sort_values(by=sort_by, ascending=ascending)

    # Format
    return (
        missing_summary
        .style
        .format({
            "Missing Count": "{:,.0f}",
            "Missing Percentage": "{:.2f}%"
        })
        .background_gradient(cmap=cmap)
    )
```

## 19. ROC-AUC

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
```

```python
classes = [0, 1, 2]
y_test_bin = label_binarize(y_test, classes=classes)
n_classes = y_test_bin.shape[1]
```

```python
def compute_roc(model, X_test, y_test_bin):
    """
    Returns:
        fpr_dict, tpr_dict, auc_dict
    """
    # Predict probabilities
    y_score = model.predict_proba(X_test)

    fpr = {}
    tpr = {}
    roc_auc = {}

    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Macro-average AUC
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
    mean_tpr = np.zeros_like(all_fpr)

    for i in range(n_classes):
        mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

    mean_tpr /= n_classes

    fpr["macro"] = all_fpr
    tpr["macro"] = mean_tpr
    roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

    # Micro-average AUC
    fpr["micro"], tpr["micro"], _ = roc_curve(
        y_test_bin.ravel(),
        y_score.ravel()
    )
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

    return fpr, tpr, roc_auc
```

```python
models = {
    "Random Forest": model_1,
    "Logistic Regression": model_2,
    "XGBoost": model_3
}
```

```python
roc_results = {}

for name, model in models.items():
    print(f"Computing ROC for {name}...")
    roc_results[name] = compute_roc(model, X_test_encoded, y_test_bin)
```

```python
plt.figure(figsize=(10, 8))

for name in roc_results:
    fpr, tpr, aucs = roc_results[name]
    plt.plot(
        fpr["macro"],
        tpr["macro"],
        label=f"{name} (macro AUC = {aucs['macro']:.3f})",
        linewidth=2
    )

plt.plot([0, 1], [0, 1], "k--", linewidth=1)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison (Macro-AUC)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
for name in roc_results:
    fpr, tpr, aucs = roc_results[name]

    plt.figure(figsize=(10, 8))
    for i in range(n_classes):
        plt.plot(
            fpr[i], tpr[i],
            label=f"Class {i} (AUC = {aucs[i]:.3f})"
        )

    plt.plot([0, 1], [0, 1], "k--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"Class-Specific ROC Curves — {name}")
    plt.legend()
    plt.grid(True)
    plt.show()
```
